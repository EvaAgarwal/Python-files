```python
import pandas as pd
import numpy as np
df=pd.read_csv('Weather.csv')
df
```

```python
new_df=df.replace([-9999,-8888],np.NaN)
new_df
#replacing values alll together
```

```python
new_df=df.replace({
    'temperature': -8888,
    'windspeed':-9999,
    'event':'0'
},np.NaN)
new_df
#replacing of values acc. to columns
```

```python
new_df=df.replace({
    -9999: np.NaN,
    'No Event':'Sunny'
    
})
new_df
#mapping of data
```

```python
#replacing alphabets using Regex
new_df=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]',
},'',regex=True)
new_df
```
