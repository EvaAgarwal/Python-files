```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
```

```python
#scatterplot linear regression 
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
dir(sns)
```

```python
#scatter plot arguments
sns.lmplot(x='total_bill',y='tip',data=tips,
           fit_reg=False,hue='smoker')
```

```python
tips.head()
```

```python
plt.figure(figsize=(20,6))
sns.lmplot(x='total_bill',y='tip',data=tips,
           fit_reg=False,hue='sex',height=6)

plt.xlim(0,None)
plt.ylim(0,None)
```

```python
plt.figure(figsize=(20,6))
sns.set_style('whitegrid') #darkgrid
sns.boxplot(data=tips)
```

```python
plt.figure(figsize=(5,5))
sns.set_style('darkgrid')
sns.violinplot(x='total_bill',y='tip',data=tips)
```

```python
tips_type_colors = ['#78C850', #total_bill
                    '#F08030',  # tip
                   ]
```

```python
plt.figure(figsize=(5,5))
sns.violinplot(x='total_bill', y='tip', data=tips, 
               palette=tips_type_colors)
```

```python
sns.swarmplot(x='total_bill', y='tip', data=tips, 
               palette=tips_type_colors)
```

```python
#overlaying plots

plt.figure(figsize=(20,6))
sns.violinplot(x='total_bill',y='tip',data=tips,
              palette=tips_type_colors)

sns.swarmplot(x='total_bill',y='tip',data=tips,
             palette=tips_type_colors)

plt.title('Tips data')
```

```python
sns.lmplot(x='total_bill',y='tip',
          col='time',hue='smoker',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',
       hue='smoker',kind='swarm',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',
            hue='smoker',kind='violin',split=True,data=tips)
```

```python
sns.catplot(x='day',y='total_bill',
            hue='smoker',kind='bar',data=tips)
```
